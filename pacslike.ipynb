```python
import requests
import pandas as pd
```

```python
cmdf = pd.read_parquet('cm.parquet')
```

```python
def pacsnear(pacs, limit=10):
    resp = requests.get(f'http://localhost:8000/pac_id_knn?limit={limit}&pac_id={pacs}')
    return pd.DataFrame(resp.json()).rename({'pacids': 'CMTE_ID'}, axis=1).join(cmdf, on='CMTE_ID')
```
